```python
# 1. VARIABLES & DATA TYPES
print("=== VARIABLES & DATA TYPES ===")

# Basic data types
name = "Alice"           # String
age = 25                 # Integer
height = 5.6            # Float
is_student = True       # Boolean

print(f"Name: {name} (type: {type(name)})")
print(f"Age: {age} (type: {type(age)})")
print(f"Height: {height} (type: {type(height)})")
print(f"Student: {is_student} (type: {type(is_student)})")
```

```python
# 2. ARITHMETIC & COMPARISON OPERATORS
print("\n=== ARITHMETIC & COMPARISON OPERATORS ===")

a, b = 10, 3

# Arithmetic operations
print(f"Addition: {a} + {b} = {a + b}")
print(f"Subtraction: {a} - {b} = {a - b}")
print(f"Multiplication: {a} * {b} = {a * b}")
print(f"Division: {a} / {b} = {a / b}")
print(f"Floor Division: {a} // {b} = {a // b}")
print(f"Modulus: {a} % {b} = {a % b}")
print(f"Power: {a} ** {b} = {a ** b}")

# Comparison operators
print(f"\nComparison: {a} > {b} = {a > b}")
print(f"Equal: {a} == {b} = {a == b}")
print(f"Not equal: {a} != {b} = {a != b}")

# Logical operators
print(f"\nLogical: True and False = {True and False}")
print(f"Logical: True or False = {True or False}")
print(f"Logical: not True = {not True}")
```

```python
# 3. STRINGS - TEXT MANIPULATION
print("\n=== STRINGS ===")

text = "Python Programming"

# String operations
print(f"Original: {text}")
print(f"Length: {len(text)}")
print(f"Uppercase: {text.upper()}")
print(f"Lowercase: {text.lower()}")
print(f"First character: {text[0]}")
print(f"Last character: {text[-1]}")
print(f"Substring (0-6): {text[0:6]}")
print(f"Replace: {text.replace('Python', 'Java')}")
print(f"Split: {text.split()}")

# String methods
text2 = "  hello world  "
print(f"\nStrip whitespace: '{text2.strip()}'")
print(f"Count 'l': {text2.count('l')}")
print(f"Starts with 'he': {text2.strip().startswith('he')}")
```

```python
# 4. LISTS - MUTABLE COLLECTIONS
print("\n=== LISTS ===")

# Create and access lists
fruits = ["apple", "banana", "orange"]
numbers = [1, 2, 3, 4, 5]

print(f"Fruits: {fruits}")
print(f"First fruit: {fruits[0]}")
print(f"Last fruit: {fruits[-1]}")
print(f"Length: {len(fruits)}")

# List operations
fruits.append("grape")        # Add to end
fruits.insert(1, "mango")     # Insert at index
fruits.remove("banana")       # Remove item
print(f"Updated fruits: {fruits}")

# List slicing
print(f"First 2 numbers: {numbers[:2]}")
print(f"Last 2 numbers: {numbers[-2:]}")
print(f"Every 2nd number: {numbers[::2]}")

# List methods
numbers.sort()
print(f"Sorted numbers: {numbers}")
print(f"Max: {max(numbers)}, Min: {min(numbers)}, Sum: {sum(numbers)}")
```

```python
# 5. DICTIONARIES - KEY-VALUE PAIRS
print("\n=== DICTIONARIES ===")

# Create dictionary
student = {
    "name": "John",
    "age": 20,
    "grade": "A",
    "subjects": ["Math", "Science", "English"]
}

print(f"Student: {student}")
print(f"Name: {student['name']}")
print(f"Age: {student.get('age', 'Not found')}")

# Dictionary operations
student["email"] = "john@email.com"  # Add new key
student["age"] = 21                  # Update existing key
print(f"Updated student: {student}")

# Get keys and values
print(f"Keys: {list(student.keys())}")
print(f"Values: {list(student.values())}")
```

```python
# 6. CONDITIONAL STATEMENTS (if-elif-else)
print("\n=== CONDITIONAL STATEMENTS ===")

# Basic if-elif-else
score = 85

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
elif score >= 60:
    grade = "D"
else:
    grade = "F"

print(f"Score: {score}, Grade: {grade}")

# Multiple conditions
age = 25
has_license = True

if age >= 18 and has_license:
    print("Can drive")
elif age >= 16 and not has_license:
    print("Can get learner's permit")
else:
    print("Cannot drive")
```

```python
# 7. LOOPS - FOR AND WHILE
print("\n=== LOOPS ===")

# For loops
print("For loop with range:")
for i in range(3):
    print(f"  Number: {i}")

print("\nFor loop with list:")
fruits = ["apple", "banana", "orange"]
for fruit in fruits:
    print(f"  I like {fruit}")

# While loop
print("\nWhile loop:")
count = 0
while count < 3:
    print(f"  Count: {count}")
    count += 1

# Loop control (break and continue)
print("\nLoop with break and continue:")
for i in range(5):
    if i == 2:
        continue  # Skip this iteration
    if i == 4:
        break     # Exit loop
    print(f"  Number: {i}")
```

```python
# 8. FUNCTIONS - REUSABLE CODE BLOCKS
print("\n=== FUNCTIONS ===")

# Basic function
def greet(name):
    """Function to greet a person"""
    return f"Hello, {name}!"

# Function with multiple parameters
def calculate_area(length, width):
    """Calculate area of rectangle"""
    return length * width

# Function with default parameters
def greet_with_title(name, title="Mr."):
    """Greet with optional title"""
    return f"Hello, {title} {name}!"

# Function calls
print(greet("Alice"))
print(f"Area: {calculate_area(5, 3)}")
print(greet_with_title("Smith"))
print(greet_with_title("Johnson", "Dr."))
```

```python
# 9. USER INPUT & TYPE CONVERSION
print("\n=== USER INPUT & TYPE CONVERSION ===")

# Type conversion examples
number_str = "123"
number_int = int(number_str)
number_float = float(number_str)

print(f"String: {number_str} (type: {type(number_str)})")
print(f"Integer: {number_int} (type: {type(number_int)})")
print(f"Float: {number_float} (type: {type(number_float)})")

# Example with predefined values (input() commented for demo)
name = "Alice"
age = 25
height = 5.6

print(f"\nHello {name}! You are {age} years old and {height} feet tall.")

# Common type conversions
print(f"str(123): {str(123)}")
print(f"int(3.14): {int(3.14)}")
print(f"float('3.14'): {float('3.14')}")
print(f"bool(1): {bool(1)}, bool(0): {bool(0)}")
```

```python
# 10. PRACTICAL EXERCISES
print("\n=== PRACTICAL EXERCISES ===")

# Exercise 1: Calculator Function
def calculator(num1, operator, num2):
    """Simple calculator function"""
    if operator == '+':
        return num1 + num2
    elif operator == '-':
        return num1 - num2
    elif operator == '*':
        return num1 * num2
    elif operator == '/':
        return num1 / num2 if num2 != 0 else "Cannot divide by zero"
    else:
        return "Invalid operator"

print("Calculator Examples:")
print(f"10 + 5 = {calculator(10, '+', 5)}")
print(f"10 * 3 = {calculator(10, '*', 3)}")
print(f"10 / 0 = {calculator(10, '/', 0)}")

# Exercise 2: Number Guessing Game (Demo)
import random
number = random.randint(1, 5)
print(f"\nGuessing Game Demo: The number is {number}")

# Exercise 3: List Operations
print("\nList Operations:")
numbers = [3, 1, 4, 1, 5, 9, 2, 6]
print(f"Original: {numbers}")
print(f"Sorted: {sorted(numbers)}")
print(f"Max: {max(numbers)}, Min: {min(numbers)}, Sum: {sum(numbers)}")
print(f"Count of 1: {numbers.count(1)}")
print(f"Index of 4: {numbers.index(4)}")
```
